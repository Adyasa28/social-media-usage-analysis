```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("test.csv")
df.head()
```

```python
df.info()
df.describe()
```

## Dataset Description
The dataset contains 103 users with attributes such as:
- Age
- Gender
- Platform
- Daily Usage Time
- Posts per day
- Likes received
- Dominant Emotion

```python
#Data Cleaning
df.isnull().sum()
```

A. Usage Time Distribution

```python
plt.figure()
plt.hist(df['Daily_Usage_Time (minutes)'], bins=10)
plt.title("Distribution of Daily Usage Time")
plt.xlabel("Minutes")
plt.ylabel("Number of Users")
plt.show()
```

```python
plt.figure()
df['Dominant_Emotion'].value_counts().plot(kind='bar')
plt.title("Dominant Emotion Distribution")
plt.show()
```

```python
plt.figure()
sns.boxplot(x='Dominant_Emotion', y='Daily_Usage_Time (minutes)', data=df)
plt.title("Usage Time vs Emotion")
plt.show()
```

```python
plt.figure()
sns.countplot(x='Platform', hue='Dominant_Emotion', data=df)
plt.xticks(rotation=45)
plt.title("Platform vs Emotion")
plt.show()
```

## Conclusion

The analysis indicates that social media usage has a measurable relationship
with users' emotional well-being.

Users with higher daily screen time demonstrate significantly higher
engagement levels, including posting, messaging, and receiving interactions.

Emotional patterns vary across platforms, with Instagram users showing
higher happiness levels, while Twitter users show more anxiety and anger.

Although most users report a neutral emotional state, a considerable portion
experience anxiety and emotional variation, suggesting that prolonged
engagement may influence emotional intensity.

Overall, social media usage does not show purely negative effects;
rather, it amplifies emotional experiences depending on usage behavior
and platform type.
